```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_flights():
    # Load dataset
    flights = sns.load_dataset("flights")

    # --- Plot 1: Line Plot (matplotlib) ---
    yearly = flights.groupby("year")["passengers"].sum()
    plt.figure(figsize=(8, 4))
    plt.plot(yearly.index, yearly.values, marker='o')
    plt.title("Total Passengers per Year")
    plt.xlabel("Year")
    plt.ylabel("Passengers")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # --- Plot 2: Bar Plot (matplotlib) ---
    monthly = flights.groupby("month")["passengers"].mean()
    plt.figure(figsize=(8, 4))
    plt.bar(monthly.index, monthly.values, color="skyblue", edgecolor="black")
    plt.title("Average Passengers per Month")
    plt.xlabel("Month")
    plt.ylabel("Average Passengers")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # --- Plot 3: Heatmap (seaborn) ---
    pivot = flights.pivot(index="month", columns="year", values="passengers")
    plt.figure(figsize=(12, 6))
    sns.heatmap(pivot, annot=True, fmt="d", cmap="YlGnBu")
    plt.title("Monthly Passengers Heatmap")
    plt.tight_layout()
    plt.show()

    # --- Plot 4: Box Plot (seaborn) ---
    plt.figure(figsize=(10, 5))
    sns.boxplot(data=flights, x="month", y="passengers")
    plt.title("Passenger Count Distribution by Month")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # --- Plot 5: Line Plot with Hue (seaborn) ---
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=flights, x="year", y="passengers", hue="month", palette="tab10")
    plt.title("Passenger Trends by Month Over Years")
    plt.tight_layout()
    plt.show()

    # --- Plot 6: Scatter Plot (seaborn) ---
    plt.figure(figsize=(10, 5))
    sns.stripplot(data=flights, x="year", y="passengers", jitter=True, hue="month", palette="tab20", dodge=True)
    plt.title("Passenger Spread by Year and Month")
    plt.ylabel("Passengers")
    plt.xlabel("Year")
    plt.legend(title="Month", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()




plot_flights()
```
